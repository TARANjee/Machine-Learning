Tools

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train is the input variable(size in 1000sqft)
#y_train in the target(prices in 1000s of dollars)
x_train=np.array([1.0,2.0])
y_train=np.array([300,500])
```

Number of training Examples

```python
m=len(x_train)
print(f"Length is {m}")
```

```python
i=0
x_i=x_train[i]
y_i=y_train[i]
print(f"x=({i}),y=({i})=({x_i},{y_i})")
```

Ploting Data

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("House Prices")
plt.ylabel("prices in $1000's")
plt.xlabel("Size(1000) sqft")
plt.show()
```

Model Function

```python
w=100
b=100
print(f"w={w},b={b}")
```

```python
def calculate_model_output(w,b,x):
    m=x.shape
    f_wb=np.zeros(m)
    for i in range(len(x)):
        f_wb[i]=w*x[i]+b
        
    return f_wb
```

```python
temp_f_wb=calculate_model_output(w,b,x_train)
plt.plot(x_train,temp_f_wb,c='b',label='Our prediction')
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual values')
plt.title("House Prices")
plt.ylabel("prices in $1000's")
plt.xlabel("Size(1000) sqft")
plt.legend()
plt.show()
```
