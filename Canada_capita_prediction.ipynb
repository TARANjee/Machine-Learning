```python
import pandas as ps
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=ps.read_csv("../dataset/Canda_per_capita_income.csv")
```

```python
%matplotlib inline
plt.xlabel("years")
plt.ylabel("per capita income US $")
plt.plot(df.year,df.perCapitaIncomeUSD,color='red')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.perCapitaIncomeUSD)
```

```python
# in year 2020 capita income in canada
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income US $')
plt.plot(df.year,df.perCapitaIncomeUSD,color='red')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
